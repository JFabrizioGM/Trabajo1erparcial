```
num=int(input("ingrese el número: "))
factorial=1
for i in range(num):
  factorial=factorial*num
  num-=1
print("el factorial es:",factorial)

```

**PROBLEMA 10** Escribir una función que calcule el área de un circulo  y otra que calcule el volumen de un cilindro utilizando la primera función.

```
print("Ingrese el radio del cirulo") 
r=float(input())
pi=3.1416
A=(pi*(r*r))
print("El área del circulo es: ",A)
h=float(input("ingrese la altura del cilindro: "))
V=A*h
print("El volumen del cilidro es: ",V)
```
